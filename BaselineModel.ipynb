```python
from sklearn.utils import shuffle
import pandas as pd
import numpy as np
import cv2
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
import tensorflow as tf
from tensorflow  import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Softmax, Conv2D, MaxPooling2D,BatchNormalization,Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report,confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder
```

```python
# load train images from 0 --> Fist 1 --> hi 2 ---> peace

#loading train_images from fist class
X_train = []
for i in range(0, 2000):
    name = 'Data/Train/fist/fist_' + str(i) + '.png'
    print(name)
    image = cv2.imread(name)
    print(image.shape)
    #convert to grayscale image
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_train.append(gray_image.reshape(240,215).flatten())

#loading train_images from hi class
for i in range(0, 2000):
    name2 = 'Data/Train/hi/hi_' + str(i) + '.png'
    image = cv2.imread(name2)
    #convert to grayscale image
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_train.append(gray_image.reshape(240, 215).flatten())
    
#Loading train_images from peace class
for i in range(0, 2000):
    name3 = 'Data/Train/peace/peace_' + str(i) + '.png'
    image = cv2.imread(name3)
    #convert to grayscale image
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_train.append(gray_image.reshape(240, 215).flatten())
```

```python
X_test = []

#like training images, we loop over test images and store them in an array

#fist images
for i in range(2000, 3000):
    image = cv2.imread('Data/Test/fist/fist_' + str(i) + '.png')
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_test.append(gray_image.reshape(240, 215).flatten())

#hi images
for i in range(2000, 3000):
    image = cv2.imread('Data/Test/hi/hi_' + str(i) + '.png')
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_test.append(gray_image.reshape(240, 215).flatten())
    
#peace images
for i in range(2000, 3000):
    image = cv2.imread('Data/Test/peace/peace_' + str(i) + '.png')
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    X_test.append(gray_image.reshape(240, 215).flatten())
    
X_test= np.array(X_test)
X_train = np.array(X_train)
```

```python
#we are going to loop over range (0, len(train_images/test_images)) and store the corresponding labels
train_labels = []


#train labels for each class 0,1,2
for i in range(0, 2000):
    train_labels.append([0])

for i in range(0, 2000):
    train_labels.append([1])

for i in range(0, 2000):
    train_labels.append([2])
    
train_labels = np.array(train_labels)
train_labels = train_labels.reshape((6000,))
Y_train = pd.get_dummies(train_labels)


test_labels = []
#test labels for each class 0,1,2
for i in range(0, 1000):
    test_labels.append([0])

for i in range(0, 1000):
    test_labels.append([1])

for i in range(0, 1000):
    test_labels.append([2])
    

test_labels = np.array(test_labels)
test_labels = test_labels.reshape((3000,))
Y_test = pd.get_dummies(test_labels)
    

```

```python
clf = clf = MLPClassifier(random_state=1, early_stopping=True).fit(X_train, Y_train)
```

```python
clf.score(X_test,Y_test)
```

```python
plt.plot(clf.loss_curve_)
plt.plot(clf.validation_scores_)
plt.ylabel('loss', fontsize = 8)
plt.xlabel('epoch', fontsize = 8)
           
plt.savefig('baseline_model_losscurve')
```

```python
import pickle
# save the classifier
with open('baselineModel.pkl', 'wb') as fid:
    pickle.dump(clf, fid)    
```
